```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv("tracks.csv")
```

```python
df_tracks.head()
```

```python
#to check sum of the null values

pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#finding top 10 least popular songs

df_least_pop = df_tracks.sort_values('popularity', ascending = True).head(10)
df_least_pop
```

```python
#To find the properties of the dataset

df_tracks.describe().transpose()
```

```python
#top 10 songs with greater than 90 popularity

df_high_pop = df_tracks.query('popularity > 90' , inplace = False).sort_values('popularity', ascending = False).head(10)
df_high_pop
```

```python
df_tracks.set_index('release_date', inplace = True)
```

```python
df_tracks.index = pd.to_datetime(df_tracks.index)
```

```python
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[12]
```

```python
df_tracks['duration'] = df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace  = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
#the Pearson correlation method is the most common method to use for numerical variables; it assigns a value between − 1 and 1, where 0 is no correlation, 1 is total positive correlation, and − 1 is total negative correlation. This is interpreted as follows: a correlation value of 0.7 between two variables would indicate that a significant and positive relationship exists between the two. A positive correlation signifies that if variable A goes up, then B will also go up, whereas if the value of the correlation is negative, then if A increases, B decreases.
corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis = 1).corr(method='pearson')
```

```python
plt.figure(figsize = (14, 6))
```

```python
heatmap = sns.heatmap(corr_df, annot = True, fmt = '.1g', vmin = -1, vmax = 1, center = 0, cmap = 'inferno', linewidth = 1, linecolor = 'Black')
heatmap.set_title("correleation between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title("Duration of songs in different genre")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(x = 'genre' , y = 'duration_ms' , data = df_genre)
plt.xlabel("Genre")
plt.ylabel("Duration In Ms")
```

```python
sns.set_style(style = 'darkgrid')
plt.figure(figsize = (10, 5))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y = 'genre', x = 'popularity', data = famous).set(title = 'top 5 popular genres')
```
